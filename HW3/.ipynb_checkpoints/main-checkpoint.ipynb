Consider the given $8\times8$ board game. There are two players in this game and they try to block each other and limit their valid moves by making walls. Their moves are limited to the 8 cells (or less when they are around the edges) beside them and the walls will start to appear only when a player steps into a random generated yellow cell and the wall will continue to be built until the player reaches the next yellow cell. The players start the game at $(0, 0)$ and $(7, 7)$ and they keep playing until one of them loses. There are multiple ways for a player to win the match:

- If a player steps into a wall or out of the board, the player loses the match!
- If the players reach the maximum moves the game finishes and the winner would be chosen based on the maximum valid move they have left plus the number of walls they have made and "No more moves" massege will appear on terminal.
- If the getMove() function does not return any value after 2 seconds, the player loses the match and "Time limit exceeded" massage will appear on terminal.

```python
import time
import copy
import sys


class TimeOutException(Exception):
    pass


class MinimaxPlayer(Player):
    def __init__(self, col, x, y, max_depth=4):
        super().__init__(col, x, y)
        self.max_depth = max_depth
        self.start_time = None

    @staticmethod
    def get_possible_moves(x, y, board):
        all_moves = [IntPair(x + i, y + j) for i in [-1, 0, 1] for j in [-1, 0, 1] if not i == j == 0]
        result = list(filter(
            lambda move: 0 <= move.x < board.getSize() and 0 <= move.y < board.getSize()
                         and board.getCell(move.x, move.y).getColor() not in (1, 2) and not (
                    move.x == x and move.y == y), all_moves))
        return result

    @staticmethod
    def is_terminal(x, y, board, depth, max_depth):
        # check depth
        if depth > max_depth:
            return True
        if not any(MinimaxPlayer.get_possible_moves(x, y, board)):
            return True
        return False

    @staticmethod
    def least_distance_to_yellow(x, y, board):
        yellow_cells = list(
            filter(
                lambda cell: cell.getId() != -1,
                [board.getCell(i, j) for i in range(board.getSize()) for j in range(board.getSize())]
            )
        )
        min_dist = sys.maxsize
        for c in yellow_cells:
            min_dist = min(min_dist, abs(c.getX() - x) + abs(c.getY() - y))
        return min_dist

    def get_utility(self, board):
        util = board.getScore(self.getCol()) - board.getScore(3 - self.getCol())
        return util

    def min_value(self, board, alpha, beta, depth):
        color = 3 - self.getCol()
        x, y = board.getPlayerX(color), board.getPlayerY(color)

        # Check for terminal state
        if MinimaxPlayer.is_terminal(x, y, board, depth, self.max_depth):
            return IntPair(x, y), self.get_utility(board)

        if time.time() - self.start_time > 2:
            raise TimeOutException()

        possible_moves = MinimaxPlayer.get_possible_moves(x, y, board)
        min_v, min_move = sys.maxsize, IntPair(x, y)
        for move in possible_moves:
            copyboard = copy.deepcopy(board)
            copyboard.move(move, color)
            v = self.max_value(copyboard, alpha, beta, depth + 1)[1]

            indent = '\t' * (depth + 1)

            if v < min_v:
                min_v = v
                min_move = move
                if v <= alpha:
                    return move, v
                if v < beta:
                    beta = v

        return min_move, min_v

    def max_value(self, board, alpha, beta, depth):
        x, y = self.getX(), self.getY()

        # Check for terminal state
        if MinimaxPlayer.is_terminal(x, y, board, depth, self.max_depth):
            return IntPair(x, y), self.get_utility(board)

        if time.time() - self.start_time > 2:
            raise TimeOutException()

        possible_moves = MinimaxPlayer.get_possible_moves(x, y, board)
        max_v, max_move = -sys.maxsize, IntPair(x, y)
        for move in possible_moves:
            copyboard = copy.deepcopy(board)
            self.setX(move.x)
            self.setY(move.y)
            copyboard.move(move, self.getCol())
            v = self.min_value(copyboard, alpha, beta, depth + 1)[1]

            indent = '\t' * (depth + 1)

            if v > max_v:
                max_v = v
                max_move = move
                if max_v >= beta:
                    return move, v
                if v > alpha:
                    alpha = v

            self.setX(x)
            self.setY(y)

        return max_move, max_v

    def getMove(self, board):
        self.start_time = time.time()
        if board.getNumberOfMoves() > board.maxNumberOfMoves:
            return IntPair(20, 20)
        if not any(MinimaxPlayer.get_possible_moves(self.getX(), self.getY(), board)):
            return IntPair(-10, -10)
        try:
            move = self.max_value(board, -sys.maxsize, sys.maxsize, 0)[0]
            return move
        except TimeOutException:
            return IntPair(10, 10)
```

```python
p1 = MinimaxPlayer(1, 0, 0)
p2 = NaivePlayer(2, 7, 7)

g = Game(p1, p2)

number_of_matches = 1
score1, score2 = g.start(number_of_matches)

print(score1 / number_of_matches)
```
